```python
df_eav = pd.read_csv('311_Enforcement_Abandoned_Vehicles.csv')
df_sc = pd.read_csv('311_Street_Cleaning.csv')
df_san = pd.read_csv('311_Sanitation.csv')
df_hm = pd.read_csv('311_Highway_Maintenance.csv')
df_ce = pd.read_csv('311_Code_Enforcement.csv')
```

```python
number_of_eav = df_eav.groupby('precinct').count()['reason']

number_of_sc = df_sc.groupby('precinct').count()['reason']

number_of_san = df_san.groupby('precinct').count()['reason']

number_of_hm = df_hm.groupby('precinct').count()['reason']

number_of_ce = df_ce.groupby('precinct').count()['reason']


```

```python
xl = pd.ExcelFile('./census_data_for_2022_redistricting.xlsx')
xl.sheet_names
census_df = xl.parse('2010')
census_df = census_df[:276]
census_df.columns=census_df.iloc[0] 
census_df = census_df[1:]
census_df
```

```python
census_df['White ratio'] = (census_df['White alone']) / (census_df['Total Population']+1)
census_df['Black ratio'] = (census_df['Black or African American']) / (census_df['Total Population']+1)
census_df['Asian ratio'] = (census_df['Asian']) / (census_df['Total Population']+1)
census_df['Hispanic ratio'] = (census_df['Hispanic or Latino (of any Race)']) / (census_df['Total Population']+1)
```

```python
census_df.index = census_df['Ward and Precinct (Updated 2022)']
```

```python
census_df
```

```python
print('Cases for Enforcement & Abandoned Vehicles')
print('White:', census_df['White ratio'].corr(number_of_eav, method='spearman'))
print('Black:', census_df['Black ratio'].corr(number_of_eav, method='spearman'))
print('Asian:', census_df['Asian ratio'].corr(number_of_eav, method='spearman'))
print('Hispanic:', census_df['Hispanic ratio'].corr(number_of_eav, method='spearman'))

df=pd.concat([census_df['White ratio'],census_df['Black ratio'],census_df['Asian ratio'],census_df['Hispanic ratio'],number_of_eav],axis=1)
df.plot(
    kind='scatter',
    x='White ratio',
    y='reason',
    title='Number of cases for Enforcement & Abandoned Vehicles versus White people ratio'
)
df.plot(
    kind='scatter',
    x='Black ratio',
    y='reason',
    title='Number of cases for Enforcement & Abandoned Vehicles versus Black people ratio'
)
df.plot(
    kind='scatter',
    x='Asian ratio',
    y='reason',
    title='Number of cases for Enforcement & Abandoned Vehicles versus Asian people ratio'
)
df.plot(
    kind='scatter',
    x='Hispanic ratio',
    y='reason',
    title='Number of cases for Enforcement & Abandoned Vehicles versus Hispanic people ratio'
)
```
