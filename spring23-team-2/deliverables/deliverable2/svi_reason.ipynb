```python
import pandas as pd
import numpy as np

df_all_311 = pd.concat([pd.read_csv('311 Cases 2010_2014 Unrestricted.csv'),pd.read_csv('311 Cases 2015_2019 Unrestricted.csv'),pd.read_csv('311 Cases 2020_2024 Unrestricted.csv')])
```

```python
df_all_311.to_csv("311_all_case.csv",index=False)
```

```python
df_gr = df_all_311[df_all_311['REASON'] == 'General Request']
df_gr.to_csv('311_General Request1.csv', index=False)
df_san = df_all_311[df_all_311['REASON'] == 'Sanitation']
df_san.to_csv('311_Sanitation1.csv', index=False)
df_sc = df_all_311[df_all_311['REASON'] == 'Street Cleaning']
df_sc.to_csv('311_Street Cleaning1.csv', index=False)
df_eav = df_all_311[df_all_311['REASON'] == 'Enforcement & Abandoned Vehicles']
df_eav.to_csv('311_Enforcement & Abandoned Vehicles1.csv', index=False)
df_hm = df_all_311[df_all_311['REASON'] == 'Highway Maintenance']
df_hm.to_csv('311_Highway Maintenance1.csv', index=False)
df_ce = df_all_311[df_all_311['REASON'] == 'Code Enforcement']
df_ce.to_csv('311_Code Enforcement.csv', index=False)
```

```python
df_gr = pd.read_csv('311_General Request1.csv')
df_san = pd.read_csv('311_Sanitation1.csv')
df_sc = pd.read_csv('311_Street Cleaning1.csv')
df_eav = pd.read_csv('311_Enforcement & Abandoned Vehicles1.csv')
df_hm = pd.read_csv('311_Highway Maintenance1.csv')
df_ce = pd.read_csv('311_Code Enforcement.csv')
```

```python
number_of_gr = df_gr.groupby('CT_ID_10').count()['REASON']
number_of_san = df_san.groupby('CT_ID_10').count()['REASON']
number_of_sc = df_sc.groupby('CT_ID_10').count()['REASON']
number_of_eav = df_eav.groupby('CT_ID_10').count()['REASON']
number_of_hm = df_hm.groupby('CT_ID_10').count()['REASON']
number_of_ce = df_ce.groupby('CT_ID_10').count()['REASON']

number_of_gr

```

```python
df_cl=pd.read_csv('Climate_Ready_Boston_Social_Vulnerability.csv')
df_cl
```

```python
df_cl['OlderAdult ratio']=(df_cl['OlderAdult'])/(df_cl['POP100_RE']+1)
df_cl['TotChild ratio']=(df_cl['TotChild'])/(df_cl['POP100_RE']+1)
df_cl['LEP ratio']=(df_cl['LEP'])/(df_cl['POP100_RE']+1)

df_cl['Low_to_No ratio']=(df_cl['Low_to_No'])/(df_cl['POP100_RE']+1)
df_cl['MedIllnes ratio']=(df_cl['MedIllnes'])/(df_cl['POP100_RE']+1)
df_cl['Disability ratio']=(df_cl['TotDis'])/(df_cl['POP100_RE']+1)

df_cl.index = df_cl['GEOID10']
df_cl

```

```python
print('Cases for General Request')
print('OlderAdult:', df_cl['OlderAdult ratio'].corr(number_of_gr, method='spearman'))
print('TotChild:', df_cl['TotChild ratio'].corr(number_of_gr, method='spearman'))
print('LEP:', df_cl['LEP ratio'].corr(number_of_gr, method='spearman'))


df=pd.concat([df_cl['OlderAdult ratio'],df_cl['TotChild ratio'],df_cl['LEP ratio'],number_of_gr],axis=1)
df.plot(
    kind='scatter',
    x='OlderAdult ratio',
    y='REASON',
    title='Number of cases for General Request OlderAdult ratio'
)
df.plot(
    kind='scatter',
    x='TotChild ratio',
    y='REASON',
    title='Number of cases for General Request TotChild ratio'
)
df.plot(
    kind='scatter',
    x='LEP ratio',
    y='REASON',
    title='Number of cases for General Request LEP ratio'
)
```
