```python
reasons = df_311['reason'].unique()
```

```python
reason_dict = {}
```

```python
for reason in reasons:
    reason_dict[reason] = 0
```

```python
reason_dict
```

```python
import numpy as np
def f1(x):
    reason_dict = {}
    for reason in reasons:
        reason_dict[reason] = 0
    
    for y in x:
        reason_dict[y]+=1
    return np.array(list(reason_dict.values()))
        
```

```python
reason_count = gp_precinct_311['reason'].aggregate(lambda x:f1(x))
```

```python
reason_count
```

```python

from sklearn.cluster import KMeans
```

```python
white_df = census_df[['Ward and Precinct (Updated 2022)','White ratio']]
black_df = census_df[['Ward and Precinct (Updated 2022)','Black ratio']]
asian_df = census_df[['Ward and Precinct (Updated 2022)','Asian ratio']]
hispanic_df = census_df[['Ward and Precinct (Updated 2022)','Hispanic ratio']]

white_df
```

```python
kmeans_white = KMeans(n_clusters=3)

kmeans_white.fit(white_df)
print(kmeans_white.cluster_centers_)
labels_white = kmeans_white.fit_predict(white_df)
```

```python
kmeans_black = KMeans(n_clusters=3)

kmeans_black.fit(black_df)
print(kmeans_black.cluster_centers_)
labels_black = kmeans_black.fit_predict(black_df)
```

```python
kmeans_asian = KMeans(n_clusters=3)

kmeans_asian.fit(asian_df)
print(kmeans_asian.cluster_centers_)
labels_asian = kmeans_asian.fit_predict(asian_df)
```

```python
kmeans_hispanic = KMeans(n_clusters=3)

kmeans_hispanic.fit(hispanic_df)
print(kmeans_hispanic.cluster_centers_)
labels_hispanic = kmeans_hispanic.fit_predict(hispanic_df)
```
