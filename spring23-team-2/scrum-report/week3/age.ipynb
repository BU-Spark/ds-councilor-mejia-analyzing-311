```python
import pandas as pd
xl = pd.ExcelFile('./census_data_for_2022_redistricting.xlsx')
xl.sheet_names
census_df = xl.parse('2020')
census_df
```

```python
census_df.columns = census_df.iloc[0]
census_df = census_df[1:]
census_df.to_csv('./census_boston_2020.csv')
census_df = census_df[:275]
census_df
```

```python
census_df.columns
```

```python
census_df['White alone ratio,\nAge 18 and over'] = (census_df['White alone,\nAge 18 and over']) / (census_df['Total Population,\nAge 18 and over']+1)
```

```python
census_df['Black or African American ratio,\nAge 18 and over'] = (census_df['Black or African American,\nAge 18 and over']) / (census_df['Total Population,\nAge 18 and over']+1)
```

```python
census_df['Hispanic or Latino (of any Race) ratio,\nAge 18 and over'] = (census_df['Hispanic or Latino (of any Race),\nAge 18 and over']) / (census_df['Total Population,\nAge 18 and over']+1)
```

```python
census_df['Asian ratio,\nAge 18 and over'] = (census_df['Asian,\nAge 18 and over']) / (census_df['Total Population,\nAge 18 and over']+1)
```

```python
census_df['American Indian and Alaska Native ratio,\nAge 18 and over'] = (census_df['American Indian and Alaska Native,\nAge 18 and over']) / (census_df['Total Population,\nAge 18 and over']+1)
```

```python
census_df['Native Hawaiian and Other Pacific Islander ratio,\nAge 18 and over'] = (census_df['Native Hawaiian and Other Pacific Islander,\nAge 18 and over']) / (census_df['Total Population,\nAge 18 and over']+1)
```

```python
census_df['Some Other Race rato,\nAge 18 and over'] = (census_df['Some Other Race,\nAge 18 and over']) / (census_df['Total Population,\nAge 18 and over']+1)
```

```python
census_df['Two or More Non-White Races ratio,\nAge 18 and over'] = (census_df['Two or More Non-White Races,\nAge 18 and over']) / (census_df['Total Population,\nAge 18 and over']+1)
```

```python
census_df.to_csv('./census_boston_2020.csv')
```
