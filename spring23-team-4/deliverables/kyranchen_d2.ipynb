```python
svi = pd.read_csv("id_requests_index.csv")
svi["SVI_Index"] = svi["SVI_Index"].fillna(0)
svi.head()

# create figure and axis objects with subplots()
fig,ax = plt.subplots()
i = range(180)
# make a plot
ax.plot(i, 
        svi.Requests,
        color="red", 
        marker="o")
# set x-axis label
ax.set_xlabel("GEOID", fontsize = 14)
# set y-axis label
ax.set_ylabel("Requests",
              color="red",
              fontsize=14)

# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(i, svi.SVI_Index,color="blue",marker="o")
ax2.set_ylabel("svi index",color="blue",fontsize=14)
plt.show()
# save the plot as a file
fig.savefig('id_request_index.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
```
