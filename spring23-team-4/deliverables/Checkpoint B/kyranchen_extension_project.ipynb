```python
import pandas as pd
#Question 1.1
df = pd.DataFrame()
#Data are collected using SQL and Excel
df["id"] = ["10-14", "15-19", "20-24"]
df["General Request"] = [0.1334, 0.1294, 0.1894]
df["Schedule a Bulk Item Pickup"] = [0.1272, 0.0531, 0.0530]
df["Requests for Street Cleaning"] = [0.0514, 0.0520, 0.0587]
df["Street Light Outages"] = [0.0484, 0.0296, 0.0224]
df["Request of Pothole Repair"] =[0.0432, 0.0472, 0.0402]
df["Parking Enforcement"] =[0.0072, 0.0878, 0.1064]
df["Reqeuest for Snow Plowing"] = [0.0503, 0.0321, 0.0031]

df.head()
```

```python
import matplotlib.pyplot as plt
gr = df["General Request"].to_list()
sbip = df["Schedule a Bulk Item Pickup"].to_list()
rsc = df["Requests for Street Cleaning"].to_list()
slo = df["Street Light Outages"].to_list()
rpr = df["Request of Pothole Repair"].to_list()
pe = df["Parking Enforcement"].to_list()
rsp = df["Reqeuest for Snow Plowing"].to_list()
id = df["id"].to_list()

plt.figure(figsize=(15, 10), dpi=100, linewidth=2)
plt.plot(id, gr, 's-', color="b", label="General Requests")
plt.plot(id, sbip, 's-', color="r", label="Schedule a Bulk Item Pickup")
plt.plot(id, rsc, 's-', color="g", label="Requests for Street Cleaning")
plt.plot(id, slo, 's-', color="c", label="Street Light Outages")
plt.plot(id, rpr, 's-', color="m", label="Request for Pothole Repair")
plt.plot(id, pe, 's-', color="y", label="Parking Enforcement")
plt.plot(id, rsp, 's-', color="k", label="Request for Snow Plowing")


plt.xlabel("Year", fontsize=15)
plt.ylabel("Percentage of Types of Requests", fontsize=15)
plt.title("Trend of Requests of Older Adult", fontsize=20)
plt.legend(loc="best")
# plt.show()
plt.savefig("Older_adults_request_trend.png", bbox_inches="tight")
```
